## How To create a Sample DATASET using the FAKER LIBRARY

```python
%pip install faker
```

```python
from faker import Faker

fake=Faker()

print(fake.name())
print(fake.country())
print(fake.email())
```

```python
data=[]

for _ in range(10000):
    data.append({
        "Name":fake.name(),
        "Email":fake.email(),
        "Phone":fake.phone_number(),
        "DOB":fake.date_of_birth(),
        "Address":fake.address(),
        "Job Title":fake.job(),
        "Company":fake.company(),
        "credit_card":fake.credit_card_number(),
        "Country":fake.country()
    })
```

```python
data
```

```python
import pandas as pd

dataset=pd.DataFrame(data)
```

```python
dataset
```

```python
dataset.to_csv("Employee_info1.csv")
```
