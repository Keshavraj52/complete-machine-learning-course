Matplotlib is a Python library that lets you create plots, charts, and graphs to visualize data. It's a popular and widely used tool for data visualization in Python. 
What can you do with Matplotlib? 
Create static, interactive, and animated visualizations
Generate plots, histograms, bar charts, scatter plots, and more
Make statistical conclusions from data

```python
import matplotlib.pyplot as plt


x=[1,2,3,4,5]
y=[1,2,3,4,12]


## create line plot

plt.plot(x,y)

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title("basic line plot")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,12]

## customize line plot

plt.plot(x,y,color='red',linestyle=':',marker='o',linewidth=3,markersize=9)# -.,--,etc 
```

```python
## multiple plots
## sample data
x=[1,2,13,4,5]
y=[11,2,3,14,12]
z=[1,12,3,4,12]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x,y,color='green')
plt.title("plot 1 ")

plt.subplot(2,2,2)
plt.plot(z,y,color='green')
plt.title("plot 2")

plt.subplot(2,2,3)
plt.plot(x,z,color='red')
plt.title("plot 3")



plt.subplot(2,2,4)
plt.plot(z,y,color='red')
plt.title("plot 4")


```

```python
categories=['A','B','C','D','E']

values=[5,6,8,3,9]

## create a bar plot

plt.bar(categories,values,color='pink')
plt.xlabel("numbers")
plt.ylabel("number")
plt.title("bar graph")
plt.show()
```

```python
## histogram
from faker import Faker
fake=Faker()
data=[]
for _ in range(500):
    data.append(
        fake.random_int(0,1000)
    )
```

```python
import matplotlib.pyplot as plt
plt.hist(data,bins=10,edgecolor='black',color='pink')
```

```python
## create a scatter plot

x=[1,2,3,5,6]
y=[2,5,8,9,6]
plt.scatter(x,y,color='red',marker='*')
```

```python
#### pie charts


labels=['A','B','C','D','E']
sizes=[30,40,50,60,65]

colors=['gold','yellowgreen','lightcoral','lightskyblue','pink']
explode=(0.2,0,0.3,0,0)
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True,explode=explode)
```

```python
import pandas as pd

emp_info=pd.read_csv('Employeeinfo.csv')
```

```python
emp_info
```

```python

df=[]
for _ in range(10000):
    df.append(fake.random_int(10000,1000000))
```

```python
emp_info['salary']=df
```

```python
emp_info.head(5)
```

```python
emp_info.info()
```

```python
## plot total salary by country


salarybycountry=emp_info.groupby('Country')['salary'].mean()
salarybycountry=salarybycountry.reset_index()
```

```python
salarybycountry.to_csv('countery_salary4320.csv',index=False)
```

```python
salarybycountry.plot(kind='bar',color='teal')
```

```python
salarybycountry.info()
```

```python
selected_country=['India','Pakistan','United States of America','China','Russian Federation']
filter=salarybycountry[salarybycountry['Country'].isin(selected_country)]

filter.plot(x="Country", y="salary", kind="bar", color="teal", legend=False)
```
