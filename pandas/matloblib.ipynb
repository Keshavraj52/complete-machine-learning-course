## data visiulization library

Matplotlib is a Python library that lets you create plots, charts, and graphs to visualize data. It's a popular and widely used tool for data visualization in Python. 
What can you do with Matplotlib? 
Create static, interactive, and animated visualizations
Generate plots, histograms, bar charts, scatter plots, and more
Make statistical conclusions from data

```python
import matplotlib.pyplot as plt


x=[1,2,3,4,5]
y=[1,2,3,4,12]


## create line plot

plt.plot(x,y)

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title("basic line plot")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,12]

## customize line plot

plt.plot(x,y,color='red',linestyle=':',marker='o',linewidth=3,markersize=9)# -.,--,etc 
```

```python
## multiple plots
## sample data
x=[1,2,13,4,5]
y=[11,2,3,14,12]
z=[1,12,3,4,12]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x,y,color='green')
plt.title("plot 1 ")

plt.subplot(2,2,2)
plt.plot(z,y,color='green')
plt.title("plot 2")

plt.subplot(2,2,3)
plt.plot(x,z,color='red')
plt.title("plot 3")



plt.subplot(2,2,4)
plt.plot(z,y,color='red')
plt.title("plot 4")


```

```python
categories=['A','B','C','D','E']

values=[5,6,8,3,9]

## create a bar plot

plt.bar(categories,values,color='pink')
plt.xlabel("numbers")
plt.ylabel("number")
plt.title("bar graph")
plt.show()
```
