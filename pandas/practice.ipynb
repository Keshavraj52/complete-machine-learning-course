```python
from faker import Faker

fake=Faker()

print(fake.name())
```

```python
data=[]

for _ in range(3000):
    data.append({
        "name":fake.name(),
        "Email":fake.email(),
        "Phone":fake.phone_number(),
        "date_of_birth":fake.date_of_birth(),
        "Address":fake.address(),
        "job":fake.job(),
        "credit_card_no":fake.credit_card_number( ),
        "country":fake.country()
    })
```

```python
data
```

```python
import pandas as pd
dataset=pd.DataFrame(data)
```

```python
dataset
```

```python
dataset.to_csv("employyess_info.csv")
```
