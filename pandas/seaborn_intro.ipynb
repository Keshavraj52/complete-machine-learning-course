```python
import seaborn as sns

tips=sns.load_dataset('tips')
```

```python
tips
```

```python
## create a scatter plot

import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("tips vs bill")
plt.show()
```

```python
## line plot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title('line plot of total bill by size')
plt.show()
```

```python
## categorical plot
#3 bar plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar plot of categorical data')
plt.show()
```

```python
## bow plot
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title('box plot of categorical data')
plt.show()
```

```python
sns.boxenplot(x='day',y='total_bill',data=tips)
plt.title('boxen plot of categorical data')
plt.show()
```

```python
## violent plot

sns.violinplot(x='day',y='total_bill',data=tips)
plt.title('voilent graph of day and bill')
```

```python
## histogram

sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
sns.kdeplot(tips['total_bill'],shade=True)
```

```python
## pairplot
sns.pairplot(tips)
```

```python
## heatmap

corr=tips[['total_bill','tip','size']].corr()

```

```python
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
